```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Single confusion matrix
conf_matrix = np.array([[3361, 2958], [572, 17094]]).T  # Transposed for desired layout

# Labels
actual_labels = ['Positive (1)', 'Negative (0)']
predicted_labels = ['Positive (1)', 'Negative (0)']

# Plotting
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="YlGnBu",
            xticklabels=actual_labels, yticklabels=predicted_labels, cbar=False)

plt.title('Naive Bayes Confusion Matrix')
plt.xlabel('Actually')
plt.ylabel('Predicted')
plt.tight_layout()
plt.show()
```
